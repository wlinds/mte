## 🎼 Music theory - Western Chromatic Scale

```python
# Reading table for western music where A4 is tuned to 440Hz
chromatic_440hz = pd.read_html('https://pages.mtu.edu/~suits/notefreqs.html')
chromatic_440hz
```

```python
# Create df, rename columns
df440 = chromatic_440hz[1]
df440.columns = ["Key", "Frequency(Hz)", "Wavelength(cm)"]
df440.head(10)
```

```python
# same thing but tuned to 432Hz
chromatic_432hz = pd.read_html('https://pages.mtu.edu/~suits/notefreq432.html')
df432 = chromatic_432hz[1]
df432.columns = ["Key", "Frequency(Hz)", "Wavelength(cm)"]

df432.head(10)
```

```python
# same thing but tuned to 432Hz
chromatic_432hz = pd.read_html('https://pages.mtu.edu/~suits/notefreq432.html')
df432 = chromatic_432hz[1]
df432.columns = ["Key", "Frequency(Hz)", "Wavelength(cm)"]

n = 1 # increments (12 is an octave)
plt.rc('figure', figsize=(16, 9), dpi=380)

plt.plot(df440['Key'].iloc[::n], df440['Frequency(Hz)'].iloc[::n], label = "Stuttgart pitch, A4 = 440Hz")
plt.plot(df432['Key'].iloc[::n], df432['Frequency(Hz)'].iloc[::n], label = "Verdi tuning, A4 = 432Hz")
plt.title("Comparing chromatic scale frequencies of key A4")
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

Clearly the difference in tonality is very subtle. Kinda neat how the flat and sharp key labels visually resembles a piano.
