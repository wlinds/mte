```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')

a = np.array([6, -2])
b = np.array([-4, 4])

# Default plotting
#  plt.plot(a)
#  plt.plot(b)
#  plt.grid() etc
```

```python
# Variables for hex code color values
grön, röd, blå, gül = '#259645', '#E23A2E', '#1A73E8', '#FCC31E'

# Create figure and axes
fig, ax = plt.subplots()

plt.rc('figure', figsize=(16, 9), dpi=200)

# Fist argument sets origin
ax.arrow(0,0, a[0],a[1], color=grön) # Green
ax.arrow(0,0, b[0],b[1], color=röd) # Red

a_neg = np.negative(a) # Create negative mirror with numpy
b_neg = [-val for val in b] # Create negative mirror with list comprehension

ax.arrow(0,0, a_neg[0],a_neg[1], color=blå)
ax.arrow(0,0, b_neg[0],b_neg[1], color=gül)

# set position at origin visually and define bounding area
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

This shows four vectors pointing in four different directions within $\mathbb{R^2}$, with their origin set to 0. Lets connect the lines:

```python
# Create figure and axes
fig, ax = plt.subplots()

# using human to count:

# from x = -6, y = 2 to x = 4, y = -4
ax.arrow(-6,2,10,-6)

# from x = 4, y = -4 to x = 6, y = -2
ax.arrow(4,-4,2,2)

# from x = 6, y = -2 to x = -4, y = -4
ax.arrow(6,-2, -10,6)

# from x = -4, y= -4 to x = -6, y = -4
ax.arrow(-4,4, -2,-2)


# ---- Same as above cell ---- #

plt.rc('figure', figsize=(16, 9), dpi=200)

# Fist argument sets origin
ax.arrow(0,0, a[0],a[1], color=grön) # Green
ax.arrow(0,0, b[0],b[1], color=röd) # Red

a_neg = np.negative(a) # Create negative mirror with numpy
b_neg = [-val for val in b] # Create negative mirror with list comprehension

ax.arrow(0,0, a_neg[0],a_neg[1], color=blå)
ax.arrow(0,0, b_neg[0],b_neg[1], color=gül)

# set position at origin visually and define bounding area
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# ---- / ---- #

# set limit on spines
plt.xlim(-8,8)
plt.ylim(-8,8)

plt.show()
```

```python
# Creating the same shape using a single vector
fig, ax = plt.subplots()
plt.rc('figure', figsize=(16, 9), dpi=200)

arr = np.array([[-4,4],[6, -2],[4, -4],[-6,2],[-4,4]])

# Split coodinates in array
x = [x[0] for x in arr]
y = [y[1] for y in arr]


# Create copy of coordinate pair, but invert ys
x2 = [x[0] for x in arr]
y2 = np.invert([y[1] for y in arr])


plt.xlim(-10,10)
plt.ylim(-6,6)

plt.plot(x, y)
plt.plot(x2, y2)

plt.grid()
plt.show()
```

```python
# drawing a visual representation of a cube in 2D space (not mathematically a cube)
yt = '#FFFFFF'

fig, ax = plt.subplots()
plt.rc('figure', figsize=(1,1), dpi=200)

# define the coordinates of the square
x = np.array([0, 5, 5, 0, 0])
y = np.array([0, 0, 5, 5, 0])

# plot the square
ax.plot(x, y, color=yt)
ax.plot(x+2,y+2, color=yt)

# plot the lines connecting the squares
plt.plot([0, 2], [0, 2], color=yt)
plt.plot([5, 7], [0, 2], color=yt)
plt.plot([5, 7], [5, 7], color=yt)
plt.plot([0, 2], [5, 7], color=yt)

ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.axis('off')
plt.show()
```

The central problem in epistemology is that of explaining *how* perception could give us knowledge or justified belief about an external world.
